# AdaBoost 代码实现

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
class AdaBoost0:
    def __init__(self, n_estimators=50):
        # 初始化参数
        self.n_estimators = n_estimators # 估计器数量 
        self.models = [] #  基学习器列表
        self.alphas = []# 权重列表

    def fit(self, X, y):
        # 样本数量
        num = X.shape[0]
        # 初始化数据集权重列表
        self.weights = np.ones(num) / num # 初始化权重为1/N

        for t in range(self.n_estimators):
            #创建决策树分类器，max_depth=1表示使用决策树桩
            model = DecisionTreeClassifier(max_depth=1)# 基学习器为决策树
            model.fit(X, y, sample_weight=self.weights)# 训练基学习器
            y_pred = model.predict(X)

            # 计算分类误差率e_m
            e_m = np.sum(self.weights * (y != y_pred))

            # 计算基学习器权重alpha
            alpha_m = 0.5 * np.log((1 - e_m) / e_m)

            #更新权重分布，使得分类错误的样本权重增加，分类正确的样本权重减少，最终权重是一个概率分布
            self.weights = self.weights * np.exp(-alpha_m * y * y_pred)
            self.weights = self.weights / np.sum(self.weights)# 规范化因子Z_m=Z_m=np.sum(self.weights)

            # 保存基学习器和权重
            self.models.append(model)
            self.alphas.append(alpha_m)
            print('Estimator: {}, Error: {}'.format(t, e_m))


    def predict(self, X):
        weak_pred = np.zeros(X.shape[0]) 
        for model, alpha in zip(self.models, self.alphas):
            
            weak_pred += alpha * model.predict(X) # 对每个决策器的预测结果乘以对应的权重
            
        return np.sign(weak_pred) # 预测结果为1或-1
            

```

```python
# 生成示例数据
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=500, n_features=20, n_informative=10, random_state=42)

# 将标签从 {0, 1} 转换为 {-1, 1}，因为 AdaBoost 是基于 -1 和 1 的分类标签
y = 2 * y - 1

# 实例化 AdaBoost 模型
ada_boost = AdaBoost0(n_estimators=50)

# 训练 AdaBoost 
ada_boost.fit(X, y)

# 预测结果
y_pred = ada_boost.predict(X)

# 计算准确率
accuracy = accuracy_score(y, y_pred)
print(f"AdaBoost准确率: {accuracy * 100:.2f}%")
```
